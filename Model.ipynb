```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
trainDataGen = ImageDataGenerator(
rotation_range = 5,
width_shift_range = 0.1,
height_shift_range = 0.1,
rescale = 1.0/255,
shear_range = 0.2,
zoom_range = 0.2,
horizontal_flip = False)

trainGenerator = trainDataGen.flow_from_directory(
"/home/superalkalinedroid/Documents/DevanagariHandwrittenCharacterDataset/Train",
target_size = (32,32),
batch_size = 32,
color_mode = "grayscale",
class_mode = "categorical")

test_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = test_datagen.flow_from_directory(
"/home/superalkalinedroid/Documents/DevanagariHandwrittenCharacterDataset/Test",
target_size=(32,32),
batch_size=32,
color_mode = "grayscale",
class_mode= 'categorical')
```

```python
from tensorflow.keras import optimizers
from tensorflow.keras import models
from tensorflow.keras import layers
imageSize=32
classifier= models.Sequential() 

classifier.add(layers.Conv2D(64, (3, 3), input_shape = (imageSize, imageSize, 1), activation = 'relu'))
classifier.add(layers.MaxPooling2D(pool_size = (2, 2)))

classifier.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
classifier.add(layers.MaxPooling2D(pool_size = (2, 2)))

classifier.add(layers.Conv2D(64, (3, 3), activation = 'relu')) 
classifier.add(layers.MaxPooling2D(pool_size = (2, 2)))

classifier.add(layers.Flatten())
```

```python
classifier.add(layers.Dense(128, activation='relu'))
classifier.add(layers.Dense(64, activation='relu'))
classifier.add(layers.Dense(46, activation='softmax'))
```

```python
classifier.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
classifier.fit(trainGenerator, epochs = 10, steps_per_epoch = 2444, validation_data = validation_generator, validation_steps = 432)
```

```python
classifier.save('handwriting_model.h5')
```
